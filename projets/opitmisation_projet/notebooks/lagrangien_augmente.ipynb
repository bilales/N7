## Interprétation

 1. Commenter les résultats obtenus, en étudiant notamment les valeurs en sortie de $\lambda_k$ et $\mu_k$.
 2. Étudier l'influence du paramètre $\tau$ dans la performance de l'algorithme. Pour cela Vous réaliserez des tests numériques.

```julia
using LinearAlgebra
using Markdown


include("../src/Lagrangien_Augmente.jl")
include("../test/fonctions_de_tests.jl")




fct1(x) = 2 * (x[1] + x[2] + x[3] - 3) ^ 2 + (x[1] - x[2]) ^ 2 + (x[2] - x[3]) ^ 2
grad_fct1(x) = [4*(x[1]+x[2]+x[3]-3)+2*(x[1]-x[2]);4*(x[1]+x[2]+x[3]-3)-2*(x[1]-x[2])+2*(x[2]-x[3]);4*(x[1]+x[2]+x[3]-3)-2*(x[2]-x[3])]
hess_fct1(x) = [6 2 4;2 8 2;4 2 6]

contrainte1(x) = x[1] + x[3] - 1
grad_contrainte1(x) = [1; 0; 1]
hess_contrainte1(x) = zeros(3, 3)

sol_exacte = [0.5; 1.25; 0.5]
x0 = [14; -11; 3]

# Influence de mu 
options_mu= [
    [1e-2, 1e-5, 1000, 2, 100, 2],
    [1e-2, 1e-5, 1000, 2, 4, 2]
]

for opts in options_mu
    xmin, f_min, flag, nb_iters, muks, lambdaks = lagrangien_augmente(
        fct1, grad_fct1, hess_fct1, contrainte1, grad_contrainte1, hess_contrainte1, x0,
        max_iter=5, tol_abs=opts[1], tol_rel=opts[2], λ0=opts[4], μ0=opts[5], τ=opts[6], algo_noc="rc-cauchy"
    )
    println("influence de mu")
    afficher_resultats("Lagrangien Augmente", "f1", x0, xmin, f_min, flag, nb_iters, sol_exacte)
    println("λk: ", lambdaks, "\nμk: ", muks, "\n")
end

options_lambda = [
    [1e-2, 1e-5, 1000, 2, 100, 2],
    [1e-2, 1e-5, 1000, 60, 100, 2]
]

for opts in options_lambda
    xmin, f_min, flag, nb_iters, muks, lambdaks = lagrangien_augmente(
        fct1, grad_fct1, hess_fct1, contrainte1, grad_contrainte1, hess_contrainte1, x0,
        max_iter=5, tol_abs=opts[1], tol_rel=opts[2], λ0=opts[4], μ0=opts[5], τ=opts[6], algo_noc="rc-cauchy"
    )
    println("influence de lambda")
    afficher_resultats("Lagrangien Augmente", "f1", x0, xmin, f_min, flag, nb_iters, sol_exacte)
    println("λk: ", lambdaks, "\nμk: ", muks, "\n")
end


# Influence de tau (τ)
options_tau = [
    [1e-2, 1e-5, 1000, 2, 100, 1],
    [1e-2, 1e-5, 1000, 2, 100, 100]
]

for opts in options_tau
    xmin, f_min, flag, nb_iters, muks, lambdaks = lagrangien_augmente(
        fct1, grad_fct1, hess_fct1, contrainte1, grad_contrainte1, hess_contrainte1, x0,
        max_iter=5, tol_abs=opts[1], tol_rel=opts[2], λ0=opts[4], μ0=opts[5], τ=opts[6], algo_noc="rc-cauchy"
    )
    println("influence de tau")
    afficher_resultats("Lagrangien Augmente", "f1", x0, xmin, f_min, flag, nb_iters, sol_exacte)
    println("λk: ", lambdaks, "\nμk: ", muks, "\n")
end
```

Les fonctions et contraintes utilisées ne relèvent sans doute pas assez l'influence des paramètres, l'algorithme choisi
est celui de region de confiance avec pas de cauchy


1)

Dans les résultats obtenus de l'algorithme, les valeurs de lambdak semblent fluctuer et s'ajuster au fur et à mesure que l'algorithme progresse. 
Si lambdak se stabilise autour d'une certaine valeur, cela peut indiquer que l'algorithme a trouvé un équilibre pour satisfaire les contraintes. 

D'un autre côté, les valeurs de muk augmentent ou restent stables dans certains cas. 
Une augmentation dans muk suggère que l'algorithme renforce la pénalité pour les contraintes non satisfaites, 
ce qui peut aider à guider la recherche vers une région viable de l'espace des solutions. 


2)
lorsque tau est faible, muk augmente lentement, ce qui donne à l'algorithme plus de flexibilité pour explorer l'espace des solutions avant de renforcer la pénalité. Cela peut être utile si l'espace des solutions a de nombreuses contraintes difficiles à satisfaire.

lorsque tau est grand, muk augmente rapidement, ce qui force l'algorithme à satisfaire rapidement les contraintes. Cela peut accélérer la convergence vers une solution réalisable, mais il y a un risque de convergence prématurée vers un minimum local non optimal si les contraintes sont trop strictement appliquées.
